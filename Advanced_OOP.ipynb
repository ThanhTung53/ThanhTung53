```python
class Person():
    name: str
    phone_number: str
    email_address: str

    def __init__(self, name, phone_number, email_address) -> None:
        super().__init__()
        self.name = name
        self.phone_number = phone_number
        self.email_address = email_address

    def outputPerson(self) -> str:
        result = "Name: " + self.name + "; Phone number: " + self.phone_number + "; Email address: " + self.email_address
        return result

class Student(Person):
    student_number: str
    average_mark: str

    def __init__(self, name, phone_number, email_address, student_number, average_mark) -> None:
        Person.__init__(self, name, phone_number, email_address)
        self.student_number = student_number
        self.average_mark = average_mark

    def outputStudent(self) -> str:
        result = self.outputPerson() + "Student Number: " + self.student_number + "; Average Mark: " + self.average_mark
        return result

class Professor(Person):
    salary: int

    def __init__(self, name, phone_number, email_address, salary) -> None:
        Person.__init__(self, name, phone_number, email_address)
        self.salary = salary

    def outputProfessor(self) -> str:
        result = self.outputPerson() + "Salary: " + str(self.salary)
        return result

def NhapN(name):
    n = "Input the number of " + str(name)
    x = int(input(n))
    while (x <= 0):
        print("Incorrectly, please try again: ")
        x = int(input("Input again: "))
    return x

def Nhap1PS():
    name = input("Person Name: ")
    phone_number = input("Person number: ")
    email_address = input("Person email: ")
    x = Person(name, phone_number, email_address)
    return x

def Nhap1SV():
    name =  input("Student name: ")
    phone_number = input("Student number: ")
    email_address = input("Student email: ")
    student_number = input("Student ID: ")
    average_mark = input("GPA: ")
    x = Student(name, phone_number, email_address, student_number, average_mark)
    return x 

def Nhap1GV():
    name =  input("Professor name: ")
    phone_number = input("Professor number: ")
    email_address = input("Professor email: ")
    salary = input("Salary: ")
    x = Professor(name, phone_number, email_address, salary)
    return x

def NhapPS(x):
    result = []
    for i in range(x): 
        print("Input information of person: ", i + 1)
        result.append(Nhap1PS())
    return result 

def NhapSV(x):
    result = [] #result = list
    for i in range(x): # x = số sinh viên
        print("Input information of student: ", i + 1)
        result.append(Nhap1SV())
    return result #NhapSV() = danh sách gồm n đối tượng sinh viên

def NhapGV(x):
    result = []
    for i in range(x):
        print("Input information of professor: ", i + 1)
        result.append(Nhap1GV())
    return result

def XuatSV(result):
    for i in result: #result = list danh sach SV
        print(i.outputStudent())

def XuatGV(result):
    for i in result:
        print(i.outputProfessor())

def XuatPS(result):
    for i in result:
        print(i.outputPerson())

def SapxepPS(result):
    for i in range(len(result)):
        for j in range(i + 1, len(result)):
            if(result[i].name > result[j].name):
                result[i], result[j] = result[j], result[i]
    return result

def SapxepSV(result):
    for i in range(len(result)):
        for j in range(i + 1, len(result)): 
            if(result[i].averagerank > result[j].averagerank):
                result[i], result[j] = result[j], result[i]
    return result

def SapxepGV(result):
    for i in range(len(result)):
        for j in range(i + 1,len(result)):
            if(result[i].salary < result[j].salary):
                result[i], result[j] = result[j], result[i]
    return result

def InVaoFilePS(result):
    with open("PS.txt", "wb+") as f:
        for i in result:
            pickle.dump(i.outputPerson(), f)

def InVaoFileSV(result):
    f = open("STD.txt", "wb+")
    for i in result: 
        pickle.dump(i.outputStudent(), f)
    f.close()

def InVaoFileGV(result):
    with open("PFS.txt", "wb+") as f:
        for i in result:
            pickle.dump(i.outputProfessor(), f)

def DocFile(tenfile, x, name):
    with open(tenfile, "rb+") as sf:
        for i in range(x):
            print(name, i+1, ":")
            print(pickle.load(sf))

import pickle

#Hàm chính
def main():
    dsps = []
    dssv = []
    dsgv = []
    sps = NhapN("Person: ") 
    ssv = NhapN("Student: ")
    sgv = NhapN("Professor: ")

    #Người
    dsps = NhapPS(sps)
    #XuatPS(sps)

    #Sinh Viên
    dssv = NhapSV(ssv)
    #XuatSV(dssv)

    #Giảng Viên
    dsgv = NhapGV(sgv)
    #XuatGV(dsgv)

    #Sapxep
    dsps = SapxepPS(dsps)
    dssv = SapxepSV(dssv)
    dsgv = SapxepGV(dsgv)

    #InFile
    InVaoFilePS(dsps)
    InVaoFileSV(dssv)
    InVaoFileGV(dsgv)

    #Docfile
    print("Person:")
    DocFile("PS.txt", sps, "Person")
    print("Student:")
    DocFile("STD.txt", ssv, "Student")
    print("Professor:")
    DocFile("PFS.txt", sgv, "Professor")

if __name__ == "__main__":
    main()
```
